## 問題28

```python
basic_info_dic={}
basic_infos = re.findall(r'\|(.*?) = (.*)',answer)
for basic_info in basic_infos:
    tmp=re.sub(r"''+","",basic_info[1])
    tmp=re.sub(r"\[\[[^\|\[\]]*?\#[^\|\[\]]*?\|([^\|\[\]]*?)\]\]",r"\1",tmp)
    tmp=re.sub(r"\[\[[^\|\[\]]*?\|(\{.*?\}|[^\{\}\|\[\]]*?)\]\]",r"\1",tmp)
    tmp=re.sub(r"\[\[([^\|\[\]]*?)\]\]",r"\1",tmp)
    tmp=re.sub(r"\{\{.*?\|.*?\|.*\|([^\}]*?)\}\}",r"\1",tmp)
    tmp=re.sub(r"\{\{.*?\|.*\|([^\}]*?)\}\}",r"\1",tmp)
    tmp=re.sub(r"\<ref\>.*\</ref\>","",tmp)
    tmp=re.sub(r"\<ref.*\>","",tmp)
    tmp=re.sub(r"\[\[ファイル:.*?\|.*?\|(.*)\]\]",r"\1",tmp)
    tmp=re.sub(r"\{\{0\}\}","",tmp)
    tmp=re.sub(r"\<br \/\>","",tmp)
    basic_info_dic.setdefault(basic_info[0],tmp)
pprint.pprint(basic_info_dic)
```
